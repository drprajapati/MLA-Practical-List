Aim: Use K-Means Clustering algorithm for dividing students based on some parameter.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../input/fashion-mnist_train.csv')
labels = dataset['label'].values
features = dataset.drop(['label'],axis=1).values
dataset.info()
```

```python
dataset.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(features)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
y_pred = KMeans(n_clusters=2, random_state=0).fit_predict(features)
```
