Aim :Visualize data using Matplotlib and derive conclusions.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('../input/boston_housing.csv')
X = dataset['age'].values
y = dataset['rm'].values
```

```python
plt.scatter(X,y,color = "red", alpha=0.5, marker=r'$\star$')
plt.title('Boston House Prediction')
plt.xlabel('Proportion of owner-occupied units built prior to 1940.')
plt.ylabel('Average number of rooms per dwelling.')
plt.show()
```
