```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../input/fashion-mnist_train.csv')
labels = dataset['label'].values
features = dataset.drop(['label'],axis=1).values
dataset.info()
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(20, 10), random_state=1)
clf.fit(features,labels)
```

```python
test_dataset = pd.read_csv('../input/fashion-mnist_test.csv')
test_labels = test_dataset['label'].values
test_features = test_dataset.drop(['label'],axis = 1).values
y_pred =clf.predict(test_features)
```

```python
print("Training set score: %f" % clf.score(features, labels))
print("Test set score: %f" % clf.score(test_features, test_labels))
```
